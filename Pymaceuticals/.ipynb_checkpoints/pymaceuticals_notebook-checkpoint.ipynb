## Observations and Insights

The following is a summary of three findings based on the analysis within this notebook:

1. Drug regimen tumor volume: Based on the data shown in the box plots, we can see that the there are two out of four drug treatments that were more successful in terms of final tumor size. Capomulin and Ramicane had final tumor volumes of about 39 mm3 and 36 mm3. Comparing this to the results for Infubinol and Ceftamin’s final tumor volumsn of about 60 mm3 for both, it is clear that Capomulin and Ramicane would be the drug treatments of choice when it comes to reducing tumor size.

2. Capulin’s treatment of mouse I509: The treatment that is displayed in the line graph of mouse I509 shows the tumor volume across the time point or days that the measurements took place. It clear that the drug treatment was successful in reducing tumor size. It appears that drug takes about 20 days until the tumor begins to reduce in size. After 20 days, the tumor dramatically decreases in size until about day 35. Upon day 35, we see a slight increase in tumor size, but this may be due to reasons unclear within our data. One hypothesis could be that the mouse had grown in size during that time period and therefore tumor size grew along with it. This was an analysis that was conducted in the scatter plot at the end of the analysis.

3. Tumor size and mouse weight: As discussed above, I hypothesized that mouse weight may be related to the tumor volume. In the scatter plots at the end of the analysis, we look at the correlation between these two variables. Based on the scatter plot alone, we can see that there is some relation between weight and tumor size. As the mouse weight increases, so too did the tumor volume. Based on Pearson's correlation coefficient is the test, we find that the correlation between mouse weight and tumor size has a value of 0.84. The closer this value is to 1, means a very strong correlation. Therefore, based on our findings we can say that there is a correlation between mouse weight and tumor size.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.
merge_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_mouse_id = merge_df.loc[merge_df.duplicated(subset=['Mouse ID', 'Timepoint',]),'Mouse ID'].unique()
dup_mouse_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
g989 = merge_df.loc[merge_df["Mouse ID"] == 'g989']
g989
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data =merge_df.loc[merge_df['Mouse ID']!='g989']
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
clean_data["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
drug_sum = clean_data[['Drug Regimen']].drop_duplicates()
drug_sum = drug_sum.sort_values(by = "Drug Regimen")
drug_sum = drug_sum.set_index("Drug Regimen")

# Assemble the resulting series into a single summary dataframe.
drug_sum['Mean'] = clean_data.groupby(['Drug Regimen']).mean()['Tumor Volume (mm3)'].values
drug_sum['Median'] = clean_data.groupby(['Drug Regimen']).median()['Tumor Volume (mm3)'].values
drug_sum['Variance'] = clean_data.groupby(['Drug Regimen']).var()['Tumor Volume (mm3)'].values
drug_sum['Standard Deviation'] = clean_data.groupby(['Drug Regimen']).std()['Tumor Volume (mm3)'].values
drug_sum['Standard Error of Measure'] = clean_data.groupby(['Drug Regimen']).sem()['Tumor Volume (mm3)'].values

drug_sum
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pandas.

#Creating the data frame for the bar chart
drug_bar = clean_data[['Drug Regimen']].drop_duplicates()
drug_bar['Number of Measurements Taken'] = clean_data.groupby('Drug Regimen').count()['Tumor Volume (mm3)'].values

#Plotting the bar chart
drug_bar.plot.bar('Drug Regimen','Number of Measurements Taken',alpha = 0.5)

plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
final_volume_df = clean_data.set_index('Drug Regimen').loc[['Capomulin','Ramicane','Infubinol','Ceftamin'],:].sort_values('Mouse ID')
# Start by getting the last (greatest) timepoint for each mouse
final_volume_df = final_volume_df.groupby("Mouse ID")["Timepoint"].max()
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merge_df = pd.merge(final_volume_df, clean_data, on=["Mouse ID","Timepoint"])
merge_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin','Ramicane','Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_list = []

#Setting for loop for the treatment list
for treatment in treatments:


# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
    merge_df_sorted=merge_df.loc[merge_df['Drug Regimen']==treatment].sort_values('Tumor Volume (mm3)')
    quartiles=merge_df_sorted['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lower=quartiles[.25]
    upper=quartiles[.75]
    IQR=upper-lower
    print(f'The IQR for the {treatment} regimen is {round(IQR)} ')
    # adding subset 
    tumor_volume_list.append(merge_df_sorted['Tumor Volume (mm3)'])
    # Determining outliers using upper and lower bounds
    outliers= merge_df_sorted.loc[(merge_df_sorted['Tumor Volume (mm3)']>1.5*IQR+upper) | (merge_df_sorted['Tumor Volume (mm3)']<lower-1.5*IQR)]
    print(f'There are {len(outliers)} outliers in the tumor volume for the mice on {treatment} regimen ')
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

#Generating the box plot
fig1, ax1 = plt.subplots()

#Setting the title, x and y axis
ax1.set_title('Final Tumor')
ax1.set_ylabel('Final tumor volume')
ax1.set_xlabel('Drug Regimen')

#Creating the box plots and printing to the screen
ax1.boxplot(tumor_volume_list, labels=["Capomulin","Ramicane","Infubinol","Ceftamin"])
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

#Creating a data frame for one mouse under the Capomulin regimen
line_df = clean_data.loc[clean_data["Drug Regimen"] == "Capomulin",:]
line_mouse_df = line_df.loc[line_df["Mouse ID"] == "l509",:]

#Setting the x and y axis of the graph
x_axis = line_mouse_df["Timepoint"]
y_axis = line_mouse_df["Tumor Volume (mm3)"]

#Plotting the graph, setting the title and x and y axis labels
plt.title("Capomulin treatment of mouse l509")
plt.plot(x_axis, y_axis)
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")

#Priting the plot to the notebook
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

#Creating the scatter plot with a group by function
scatter_df = line_df.groupby(["Mouse ID"]).mean()
#Plotting the scatter plot
plt.scatter(scatter_df["Weight (g)"], scatter_df["Tumor Volume (mm3)"])

#Labeling the x and y axis and setting a graph title
plt.xlabel("Weight(g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.title("Capomulin Treatment Tumor and Weight Relation")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#Setting x and y values
x_values = scatter_df["Weight (g)"]
y_values = scatter_df["Tumor Volume (mm3)"]
#Setting the linear regression model
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values,y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
#Plotting the scatter plot
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,30),fontsize=15,color="red")
#Labeling the x and y axis
plt.xlabel('Weight')
plt.ylabel('Tumor Volume (mm3)')
#Calculating and printing the correlation
correlation=round(st.pearsonr(scatter_df['Weight (g)'],scatter_df['Tumor Volume (mm3)'])[0],2)
print(f'The correlation between mouse weight and average tumor volume is {correlation}')
plt.show()

```
